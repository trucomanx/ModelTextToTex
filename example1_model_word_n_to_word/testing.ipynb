# Input directory

```python
import os

BASE_DIR    = '../../outputs/example1_word_n_to_word'

TOKENIZER_DIR = os.path.join(BASE_DIR,'tokenizer')
MODEL_DIR     = os.path.join(BASE_DIR,'model')
```

# Loading tokenizer

```python
from tensorflow.keras.preprocessing.text import tokenizer_from_json
import json

with open(os.path.join(TOKENIZER_DIR,'tokenizer.json'), 'r', encoding='utf-8') as f:
    tokenizer_json = f.read()

tokenizer = tokenizer_from_json(tokenizer_json)
total_words = len(tokenizer.word_index) + 1

print(tokenizer.word_index)
print("total_words:",total_words)
```

# Loading input shape

```python
input_shape=0
with open(os.path.join(MODEL_DIR,'input_shape.json'), 'r') as archivo_json:
    datos_cargados = json.load(archivo_json)
    input_shape = datos_cargados["input_shape"]

print("input_shape:",input_shape)
```

# Loading model

```python
import mymodules.model as mmm
import tensorflow as tf

model = mmm.get_model(total_words, input_shape)
model.load_weights(os.path.join(MODEL_DIR,'best_model.weights.h5'))  # Carregue os pesos


# model = tf.keras.models.load_model("best_model.keras")
# input_shape = model.input_shape[1]
```

# Evaluate

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

seed_text = "April is"
next_words = 9

print(0,':',seed_text)

for Id in range(next_words):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]

    token_list = pad_sequences([token_list], maxlen=input_shape, padding='pre')
    predicted  = np.argmax(model.predict(token_list,verbose=False), axis=-1)

    output_word = tokenizer.index_word[predicted[0]]

    seed_text += " " + output_word
    
    print(Id+1,':',seed_text)
```
